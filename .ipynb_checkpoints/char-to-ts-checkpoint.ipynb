```python
testData = np.array([1, 2, 3])
```

```python
def normalize( rawData ) : 
    # Q = (Q - mean(Q)) / std(Q)
    mean = np.mean( rawData ) 
    sd   = np.std( rawData )
    print ("mean : {:.4} , sd : {:.4} ".format(mean,sd))
    normFunction = lambda q: (q - mean) / sd 
    normData = normFunction(rawData)
    plt.plot(rawData)
#     plt.plot(normData)
    plt.show()



#     print( rawData ) 
#     print( normData)
```

```python
normalize( testData )    
```
